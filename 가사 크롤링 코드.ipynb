```python
import os
import pandas as pd
import requests
import time
from bs4 import BeautifulSoup
import pickle
import codecs
```

```python
os.getcwd()
```

```python
with open('G:/내 드라이브/P-SAT/2020-2/주제분석/song_id.pickle', 'rb') as f:
    code_set = pickle.load(f)
```

```python
code_list = sorted(list(code_set))
print(len(code_list))
```

```python
wd_lyric = 'G:/내 드라이브/P-SAT/2020-2/주제분석/가사'
for fold in range(len(code_list)//10):
    if not os.path.isdir(f"{wd_lyric}/lyrics_{fold}"):
        os.makedirs(f"{wd_lyric}/lyrics_{fold}")
```

```python
url_base = 'https://www.melon.com/song/detail.htm?songId='
from selenium import webdriver
driver = webdriver.Chrome('chromedriver')
for num, code in enumerate(code_list):
    url = url_base + str(code)
    driver.get(url)
    try:
        driver.find_element_by_class_name("button_more.arrow_d").click() # 가사 펼치기 클릭
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        lyrics=[]
        lyrics=soup.find("div",{'id':'d_video_summary'})
    except:
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        lyrics=[]
        lyrics=soup.find("div",{'id':'d_video_summary'})
    mola = num//1000
    path = f"{wd_lyric}/lyrics_{mola}"
    
    with codecs.open (f"{path}/lyrics_{code}.txt", "w", "utf-8") as file:
        file.write(f'{lyrics}')
    if mola%10:
        os.system('cls')
        
    print(f"음원 코드 {code} 저장 완료. 총 {num/len(code_list)*100}% 완료")
```
